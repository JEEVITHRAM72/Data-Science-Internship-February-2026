## 1: Smart Light Controller

```python
    class SmartLight:

        # Define constructor with attributes name and is_on status
        def __init__(self, name):
            self.name = name
            self.is_on = False  # default is kept off 

        # Method to turn on the light
        def turn_on(self):
            if self.is_on:
                return f"{self.name} is already On"
            else:
                self.is_on = True
                return f"{self.name} is turned ON."

        # Method to turn off the light
        def turn_off(self):
            if not self.is_on:
                return f"{self.name} is already OFF"
            else:
                self.is_on = False
                return f"{self.name} is turned OFF."
                
        # Method to display the status
        def display_status(self):
            if self.is_on:
                status = "ON"
            else:
                status = "OFF"

            return f"{self.name} is {status}"
        

# Create an object
bulb1 = SmartLight("Bedroom Light")

# Call the methods from the class
print( bulb1.turn_on())
print(bulb1.turn_on())
print(bulb1.turn_off())
print(bulb1.turn_off())
print(bulb1.display_status())
```

## 2: Employee ID Card System

```python
class Employee:

    # Define the constructor and set the attributes
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to generate ID card -> display the details
    def generate_id_card(self):
        print("_"*20)
        print("Employee ID Card")
        print("Employee Name:",self.name)
        print("Employee ID:",self.emp_id)
        print("Department:",self.department)
        print("_"*20)

        
e1 = Employee("Prem", "EMP102", "AIML")
e1.generate_id_card()
```

## 3: Mobile Contact Record

```python
class Contacts:

    # Define constructor with attributes name and contact number
    def __init__(self, name, number):
        self.name = name
        self.number = number
        print("Contact Saved")

    # Method to display the contact
    def display_info(self):
        return f"{self.name} : {self.number}"

# Object 1 created
c1 = Contacts("Prem", 123456789)
print(c1.display_info())

# Object 2 created
c2 = Contacts("Jayraj", 654987123)
print(c2.display_info())
```

## 4: Product Price Tag Generator

```python
class Products:

    # Define constructor with attributes name and price 
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        return f"Product: {self.name} \nPrice: {self.price}"

# Created object of product 1
p1 = Products("Headphones", 2499)
print(p1.display_price_tag())
```

## 5: Movie Rating Display System

```python
class Movies:

    # Define constructor with attributes movie name and ratings 
    def __init__(self, name, ratings):
        self.name = name
        self.ratings = ratings

    def movie_details(self):
        return f"Movie: {self.name} \nRating: {self.ratings}/5"

# Created object of movie 1
m1 = Movies("Inception", 4.8)
print(m1.movie_details())
```

## 6: Delivery Address Manager

```python
class Customer:

    # Define constructor with attributes customer name and delivery address 
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def delivery_details(self):
        return f"Delivery Details: \nCustomer: {self.name} \nAddress: {self.address}"

# Created object of customer 1
c1 = Customer("Suman", "Hyderabad")
print(c1.delivery_details())
```
