```python
# Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Invalid amount. Minimum is 50.")
            continue

        if amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid plan. Try again.")



# Problem 2: Inventory Reorder Alert System
def check_inventory(stock_dict):
    for product in stock_dict:
        qty = stock_dict[product]
        if qty < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")



# Problem 3: Student Result Processing System
def student_result(marks_list):
    total = 0
    for m in marks_list:
        total += m
    avg = total / len(marks_list)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"



# Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare():
    base_fare = 50

    while True:
        km = float(input("Enter distance in km: "))
        peak = input("Peak hour? (yes/no): ")

        fare = base_fare + (12 * km)

        if peak.lower() == "yes":
            fare += fare * 0.25

        print("Total Fare:", fare)

        again = input("Calculate again? (yes/no): ")
        if again.lower() != "yes":
            break



# Problem 5: Employee Attendance Eligibility Checker
def check_eligibility(attendance_list):
    present = 0
    for a in attendance_list:
        if a == "P":
            present += 1

    percent = (present / len(attendance_list)) * 100

    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"



# Problem 6: Password Strength Checker
def password_strength(pw):
    if len(pw) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False

    for ch in pw:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
```
