```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_post_engagement(l):
    t = sum(l)
    s = "Viral Post" if t >= 1000 else "Normal Engagement"
    print("Total Likes:", t)
    print("Post Status:", s)


# Problem 2: Healthcare – Medicine Stock Alert System
def medicine_stock_alert(s):
    st = "Low Stock Alert" if s < 10 else "Stock Sufficient"
    print("Medicine Stock:", s)
    print("Status:", st)


# Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rl, r):
    a = sum(rl) / len(rl)
    s = "Adequate Rainfall" if a >= r else "Inadequate Rainfall"
    print("Average Rainfall:", a)
    print("Rainfall Status:", s)


# Problem 4: Social Media – Duplicate Account Detection
def detect_duplicate_accounts(u):
    d = len(u) != len(set(u))
    print("Duplicate Accounts Found:", "Yes" if d else "No")


# Problem 5: Healthcare – Appointment Eligibility Checker
def appointment_eligibility(a):
    s = "Eligible" if a >= 18 else "Not Eligible"
    print("Patient Age:", a)
    print("Eligibility Status:", s)


# Problem 6: Agriculture – Premium Crop Price Filter
def premium_crop_filter(p):
    pr = [val for val in p if val > 2000]
    print("Premium Crops:", pr)


# Problem 7: System Monitoring – Application Health Checker
def app_health_checker(e):
    if e == 0:
        s = "Healthy"
    elif e <= 5:
        s = "Minor Issues"
    else:
        s = "Critical Issues"
    print("Error Count:", e)
    print("System Status:", s)


# Problem 8: Banking – Daily Transaction Limit Checker
def transaction_limit_checker(a):
    s = "Approved" if a <= 50000 else "Rejected"
    print("Transaction Amount:", a)
    print("Transaction Status:", s)


# Problem 9: E-Learning – Student Attendance Eligibility System
def attendance_eligibility(al):
    p = (sum(al) / len(al)) * 100
    s = "Eligible" if p >= 75 else "Not Eligible"
    print("Attendance Percentage:", p)
    print("Exam Eligibility:", s)


# Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(u):
    if u <= 100:
        b = u * 3
    elif u <= 200:
        b = 100 * 3 + (u - 100) * 5
    else:
        b = 100 * 3 + 100 * 5 + (u - 200) * 7

    if b < 500:
        us = "Low Usage"
    elif b <= 1500:
        us = "Moderate Usage"
    else:
        us = "High Usage"

    print("Total Bill:", b)
    print("Usage Status:", us)
```
