```python
#Problem 1: Unique Words in a Sentence
sentence = "This is a sample sentence, this sentence is a sample."
words = sentence.lower().replace('.', '').replace(',', '').split()
unique_words = set(words)
print(f"Original sentence: {sentence}")
print(f"Unique words: {unique_words}")
```

```python
#Problem 2: Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
#Problem 3: Find Maximum and Minimum Values
array = [23,45,66,12,63,90,2]
print(f"The minimum values are {min(array)} and the maxumum values are {max(array)}")
```

```python
#Problem 4: Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
  if price > 1000:
    count +=1
print(f"The products above 1000 is {count}")
```

```python
#Problem 5: Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
toatal_days = len(attendance)
present_value = attendance.count("P")
percentage = (present_value/toatal_days) * 100
print(
  f"The attemdence is {percentage}"
)
```

```python
#Problem 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_ph_nums = set(phone_numbers)
print(f"After removing duplicates the list is{unique_ph_nums}")
```

```python
#Problem 7: Count Character Frequency
string = "hello world"
dict = {}
for char in string:
  if char in dict:
    dict[char] += 1
  else:
    dict[char] = 1
print(dict)
```

```python
#Problem 8: Convert List to Tuple
my_list = [1, 2, 3, 4, 5]
my_tuple = tuple(my_list)
print(my_tuple)
```

```python
#Problem 9: Check if a Key Exists in a Dictionary
dict = {'a': 1, 'b': 2, 'c': 3}
key = 'b'
if key in dict:
  print(f"The key '{key}' exists in the dictionary.")
else:
  print(f"The key '{key}' is not exists in the dictionary.")
```
