## 1. Smart Parking Lot Management System

```python
def smart_parking(capacity, vehicle_logs):
    '''returns current parking, peak parking and parking status'''
    
    # Initialize with setting current and peak as 0
    current = 0
    peak = 0
    
    # Iterate through logs and increment current for IN and decrement from OUT
    for logs in vehicle_logs:
        if logs == "IN":
            current += 1
        else:
            current -= 1

        # To calculate the peak parking value = assigning value of highest current
        if peak < current:
            peak = current

    # Check  for avialability
    if current >= capacity:
        status = "Not Available"
    else:
        status = "Available"

    return current, peak, status

def main():
    capacity = 50 
    vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

    currently_parked, peak, status = smart_parking(capacity, vehicle_logs)

    print("Currently parked vehicles:",currently_parked)
    print("Parking Status:",status)
    print("Peak parking usage:",peak)

main()
```

## 2. Online Food Delivery Time Estimator 

```python
def delivery_time_estimator(distance, traffic, weather):
    '''Returns final estimated food dilivery time'''

    # Set base time 
    base_time = distance * 3

    # Storing the delays extra time value in dictionaries
    traffic_delay = {"low": 0, "medium":5, "high":10}
    weather_delay = {"clear":0, "cloudy":3, "rainy":5}

    # Calculate final total eta with inputs
    eta = base_time + traffic_delay[traffic] + weather_delay[weather]

    return eta

def main():

    # Take input for distance, traffic and weather
    distance = int(input("Enter total distance: "))
    traffic = input("How was the traffic(Low/Medium/High): ").lower()
    weather = input("How was the weather(Clear/Cloudy/Rainy): ").lower()

    eta = delivery_time_estimator(distance, traffic, weather)
    print(f"Estimated delivery time: {eta}min")

main()
```

## Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy_analyzer(total_seats, total_entries):
'''Returns occupancy and status of the show'''
    
    occupancy = round(total_entries/total_seats * 100, 1)

    # If occupancy >= 75 : almost full
    # If occupancy == 100 : Housefull
    # If occupancy < 75 : Available

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    return occupancy, status
        
def main():
    total_seats = 200
    total_entries = 150

    occupancy, status = seat_occupancy_analyzer(total_seats, total_entries)

    print(f"Occupancy: {occupancy}%")
    print("Show Status:",status)

    if occupancy >= 90:
        print("Consider Opening Additional Show")

main()
```

## 4. Cloud Server Load Classification System

```python
def server_load_classifier(readings):
    '''returns average cpu load percentage and warning status'''

    # Calculate average
    average_percentage = int(round(sum(readings) / len(readings)))

    # Applying conditions
    if average_percentage < 50:
        status = "Normal"
    elif average_percentage >= 50 and average_percentage < 80:
        status = "Warning"
    else:
        status = "Critical"

    return average_percentage, status


def main():

    # CPU Readings in %
    cpu_readings = [45, 60, 70, 85, 90]

    cpu_average, status = server_load_classifier(cpu_readings)

    print(f"Average CPU Load: {cpu_average}%")
    print("Server Status:",status)

main()
```

## 5. Smart Classroom Resource Usage Monitor

```python
def usage_monitor(usage):
    threshold_usage = 5

    overused = []

    for resource, hours in usage.items():
        if hours >= threshold_usage:
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"
        
    return overused, alert

def main():

    # Resource usage in hours
    resource_usage = { 
                    "Projector": 6, 
                    "AC": 9, 
                    "Lights": 4 
                     }

    overused, alert = usage_monitor(resource_usage)

    print(f"Overused Resources: {overused}")
    print(f"Energy Alert: {alert}")

main()
```

## 6. Online Event Registration Capacity Controller

```python
def regis_capacity_control(capacity, registrations):
    '''Returns Cornfirm bookings, waiting list count and booking status'''

    # Initialize with default values for status and wait list
    status = "Closed"
    wait_list = 0

    # Apply conditions for registrations
    if capacity <= registrations:
        booked = capacity
        wait_list = registrations - capacity
    else: 
        booked = registrations
        status = "Open"
        

    return booked, wait_list, status


def main():

    # Sample inputs
    event_capacity = 100 
    registrations = 105

    booked, wait_list, status = regis_capacity_control(event_capacity, registrations)


    print("Confirmed registration :", booked)
    print("Waitlisted Users:",wait_list)
    print("Registration Status:",status)

main()
```
